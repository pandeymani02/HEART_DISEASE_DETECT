## Splitting Dataset 

```python
df_train , df_test= train_test_split(df,test_size=0.2,random_state=42)
```

```python
len(df_train),len(df_test)
```

```python
##creating object of standard scaler
scaler=StandardScaler()

def get_features_and_target_arrays(df,numericals_cols,cat_cols,scaler):
    x_numeric_scaled=scaler.fit_transform(df[numerical_cols])   ##normalization (0,1) scale
    x_categorical=df[cat_cols].to_numpy()   ## categorical cols
    x=np.hstack((x_categorical,x_numeric_scaled))
    y=df["HeartDisease"]
    
    return x,y
    
```

```python
x_train, y_train=get_features_and_target_arrays(df_train,cat_cols,numerical_cols,scaler)
```

```python
X = df.drop("HeartDisease", axis=1)

y = df["HeartDisease"]
```

# Logistic Regression Model

```python
## Training of the model 

clf=LogisticRegression()
clf.fit(x_train,y_train)
```

```python
x_test,y_test=get_features_and_target_arrays(df_test,cat_cols,numerical_cols,scaler)
```

```python
test_pred=clf.predict(x_test)
```

```python
print("Error:",mean_squared_error(y_test,test_pred))
```

```python
print("Accuracy:",accuracy_score(y_test,test_pred))
```

```python
confusion_matrix(y_test,test_pred)
```

```python
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, test_pred):
    """
    Plots a nice looking confusion matrix using Seaborn's heatmap()
    """
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(confusion_matrix(y_test, test_pred),
                     annot=True,
                     cbar=False)
    plt.xlabel("True label")
    plt.ylabel("Predicted label")
    
    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    
plot_conf_mat(y_test, test_pred)
```

```python
# Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier(),
         " Support Vector Machine":SVC(kernel='linear', random_state=0)   }

#Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python

model_scores = fit_and_score(models=models,
                             X_train=x_train,
                             X_test=x_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=["        Accuracy"])
model_compare.T.plot.bar();
```

```python
# Save the trained model as a pickle file
with open('trained_model.pkl', 'wb') as f:
    pickle.dump(clf, f)

# Load the trained model from the pickle file
with open('trained_model.pkl', 'rb') as f:
    clf = pickle.load(f)

# Function to make prediction
def predict_heart_disease():
    # Get user inputs for required attributes
    age = float(input("Enter age: "))
    sex = int(input("Enter sex (0 for female, 1 for male): "))
    cp = int(input("Enter cp (chest pain type): "))
    trestbps = float(input("Enter trestbps (resting blood pressure): "))
    chol = float(input("Enter chol (serum cholesterol in mg/dl): "))
    fbs = int(input("Enter fbs (fasting blood sugar > 120 mg/dl): "))
    thalach = float(input("Enter thalach (maximum heart rate achieved): "))
    exang = int(input("Enter exang (exercise induced angina): "))
    oldpeak = float(input("Enter oldpeak (ST depression induced by exercise relative to rest): "))
    slope = int(input("Enter slope (the slope of the peak exercise ST segment): "))

    # Create a DataFrame to hold the user inputs
    input_data = pd.DataFrame({'age': [age], 'sex': [sex], 'cp': [cp], 'trestbps': [trestbps],
                               'chol': [chol], 'fbs': [fbs], 'thalach': [thalach],
                               'exang': [exang], 'oldpeak': [oldpeak], 'slope': [slope]})

    # Use the trained model to make prediction
    prediction = clf.predict(input_data.values)

    return prediction[0]

# Call the function to make prediction
result = predict_heart_disease()

# Print the prediction result
print("Predicted heart disease: ", result)

```
